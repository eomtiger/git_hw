# 1. 도형 만들기
아래의 명세를 읽고 Python 클래스를 활용하여 점(Point)과 사각형(Rectangle)을
표현하시오.

Point 클래스에 대한 명세는 다음과 같다.

```python
class Point():
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        pass
    
```

```python
class Rectangle:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
    
    def get_area(self):
        return abs((p1.x - p2.x) * (p1.y - p2.y))
    
    def get_perimeter(self):
        return (abs(p1.x-p2.x) + abs(p1.y - p2.y)) * 2
    
    def is_square(self):
        if abs(p1.x-p2.x) == abs(p1.y - p2.y):
            return True
        else:
            return False
```

```python
p1 = Point(4,3)
p2 = Point(7,6)
```

```python
rectangle_info = Rectangle(p1,p2)
```

```python
print(rectangle_info.get_area())
print(rectangle_info.get_perimeter())
print(rectangle_info.is_square())
```
